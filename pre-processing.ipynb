```python
from PIL import Image, ImageEnhance, ImageFilter
import os
import numpy as np
import pyheif
import random
```

### 5 phép làm giàu

```python
folder_path = '/home/tower1/GitHub/citrus/data_citrus_level/black_spot'

processed_folder_path = '/home/tower1/GitHub/citrus/data_citrus_level/augmented/black_spot_early/'
os.makedirs(processed_folder_path, exist_ok=True)

# Horizontal flip
for filename in os.listdir(folder_path):
    if filename.endswith('.jpg') or filename.endswith('.JPG'):
        image_path = os.path.join(folder_path, filename)

        image = Image.open(image_path)

        image_copy = image.copy()

        flipped_image = image_copy.transpose(Image.FLIP_LEFT_RIGHT)

        flipped_image_path = os.path.join(processed_folder_path, 'horizontal_flipped_' + filename)
        flipped_image.save(flipped_image_path)

# Vertical flip
for filename in os.listdir(folder_path):
    if filename.endswith('.jpg') or filename.endswith('.JPG'):
        image_path = os.path.join(folder_path, filename)

        image = Image.open(image_path)

        image_copy = image.copy()

        flipped_image = image_copy.transpose(Image.FLIP_TOP_BOTTOM)

        flipped_image_path = os.path.join(processed_folder_path, 'vertical_flipped_' + filename)
        flipped_image.save(flipped_image_path)

# Rotate 60
# Góc xoay
rotation_angle = 60

# Lặp qua tất cả các tệp tin trong thư mục
for filename in os.listdir(folder_path):
    # Kiểm tra đuôi file (chỉ xử lý ảnh)
    if filename.endswith('.jpg') or filename.endswith('.JPG'):
        # Đường dẫn tới ảnh gốc
        image_path = os.path.join(folder_path, filename)

        # Đọc ảnh gốc
        image = Image.open(image_path)

        # Xoay ảnh
        rotated_image = image.rotate(rotation_angle)

        # Lưu ảnh đã xoay
        rotated_image_path = os.path.join(processed_folder_path, 'rotated_60_' + filename)
        rotated_image.save(rotated_image_path)

# brightness
brightness_factor = 1.2  # Adjust brightness as needed (1.0 is the original image)

for filename in os.listdir(folder_path):
    if filename.endswith('.jpg') or filename.endswith('.JPG'):
        image_path = os.path.join(folder_path, filename)

        image = Image.open(image_path)

        # Adjust brightness
        enhancer = ImageEnhance.Brightness(image)
        adjusted_image = enhancer.enhance(brightness_factor)

        adjusted_image_path = os.path.join(processed_folder_path, 'brightness_' + filename)
        adjusted_image.save(adjusted_image_path)

# zoom
zoom_factor = 0.5  # Adjust the zoom factor as needed

for filename in os.listdir(folder_path):
    if filename.endswith('.jpg') or filename.endswith('.JPG'):
        image_path = os.path.join(folder_path, filename)

        image = Image.open(image_path)

        # Calculate new size based on the zoom factor
        width, height = image.size
        new_width = int(width * zoom_factor)
        new_height = int(height * zoom_factor)

        # Resize the image to the new size
        zoomed_image = image.resize((new_width, new_height), Image.LANCZOS)

        zoomed_image_path = os.path.join(processed_folder_path, 'zoomed_' + filename)
        zoomed_image.save(zoomed_image_path)
```
